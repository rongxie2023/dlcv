```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("合作.xlsx", 0, na_values="")
df
```

```python
df1=df.groupby(["部门","公司"]).agg({'日期':'min'})
df1
```

```python
df2=df.groupby(["部门","公司"]).agg({'日期':'max'})
df2
```

```python
df11=df1.reset_index()
df22=df2.reset_index()
df11
```

```python
df3=pd.merge(df11, df22, how='left',left_on=['部门', '公司'],right_on=['部门', '公司'])
df3
```

```python
df3["合作年限"]=df3["日期_y"].dt.year-df3["日期_x"].dt.year+1
df3
```

```python
with pd.ExcelWriter("./结果.xlsx") as writer:
    df3.to_excel(writer,encoding="utf_8_sig", index=False)
```
